# Setup

This notebook generates a CSV table summarising the `comb` and `match` count predictions for each audio file in an inference directory.

```python
import pandas as pd
import os
```

# Step 1 - Define Inference Directory
This step assumes that you have already run inference using the Voxaboxen python script.
The inference results should be located within the *inference* subfolder under the relevant model run folder.

```python
INFERENCE_DIR = "/home/admin/julia-dev/voxaboxen/projects/Lit_caerulea_experiment/Lit_caerulea_bidirectional_2/inference"


# Extract unique filenames from inference directory
filenames = [x.split('-')[0].split('_')[-1] for x in os.listdir(INFERENCE_DIR) if x.endswith('.txt')]
unique_filenames = list(set(filenames))
print(f'Found {len(unique_filenames)} unique filenames in inference directory')
```

# Step 2 - Generate Output CSV

```python
CSV_PATH = "test.csv"

# Generate test CSV
rows = []
for fn in unique_filenames:
    
    comb_file = os.path.join(INFERENCE_DIR, f'peaks_pred_{fn}-comb.txt')
    with open(comb_file, 'r') as fp:
        lines = len(fp.readlines())
    comb_count = lines - 1

    match_file = os.path.join(INFERENCE_DIR, f'peaks_pred_{fn}-match.txt')
    with open(match_file, 'r') as fp:
        lines = len(fp.readlines())
    match_count = lines - 1

    rows.append({
        'capture_id': fn,
        'comb_count': comb_count,
        'match_count': match_count,
    })

test_df = pd.DataFrame.from_records(rows)
test_df.to_csv(CSV_PATH)
print(f"Saved inference results to: {CSV_PATH}")

test_df
```
